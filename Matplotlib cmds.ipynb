```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,50)#50 middle values
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
plt.show
```

```python
plt.plot(y,z,'r-')#r defines color and - defines line If we used ., we would get dotted line
plt.xlabel("wtf")
plt.ylabel("wtf2")
plt.title("Hello")
plt.show()
```

```python
#Bar graph
lang=['english','hindi','spanish']
ppl=[1000,500,600]
plt.bar(lang,ppl)
```

```python
#piechart
plt.pie(ppl,labels=lang)
```

```python
#Scatterplot
plt.scatter(x,y,color='r')
```

```python
#3-D scatter plot
plt.scatter(x,y,z)
```
