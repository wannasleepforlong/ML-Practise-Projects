```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
diabetes_df=pd.read_csv('/content/diabetes.csv')
diabetes_df.head()
```

```python
#checking for duplicates
diabetes_df.duplicated().sum()
```

```python
#checking for null
diabetes_df.isnull().sum()
```

```python
#Separating featres and labels
X=diabetes_df.drop(columns=['Outcome','SkinThickness','Insulin'],	axis=1)
Y=diabetes_df['Outcome']
```

```python
#Splitting
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
#Standardizing the data
scaler=StandardScaler()
scaler.fit(X_train)
X_train_std=scaler.transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
model=Sequential()
model.add(Dense(8,activation='relu',input_shape=(6,)))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
history=model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=200)
```
