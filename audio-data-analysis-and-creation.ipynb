# Mel Spectrogram  Conversion Example

The color scale is red-green-blue, where blue corresponds to low amplitudes, or “loudness,” and red corresponds to high amplitudes. 

The Y and X axis represent time and frequency respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import librosa
```

```python
specto=[]
def melspec_convert(fpath):
    y, sr = librosa.load(fpath)
    mels=librosa.feature.melspectrogram(y=y, sr=sr)
    specto.append([fpath,y,sr])
    librosa.display.specshow(librosa.power_to_db(mels, ref=np.max), sr=sr, x_axis='time', y_axis='mel')
    plt.title('Mel spectrogram')
    plt.show()    
```

```python
melspec_convert('/kaggle/input/gtzan-dataset-music-genre-classification/Data/genres_original/disco/disco.00071.wav')
```
