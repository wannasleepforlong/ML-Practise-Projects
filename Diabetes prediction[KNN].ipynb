```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Load the diabetes dataset
diabetes = pd.read_csv('/content/diabetes.csv')

# Handling missing values by replacing zeros with column means
c = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in c:
    mean = diabetes[col].replace(0, np.NaN).mean()
    diabetes[col] = diabetes[col].replace(0, mean)

# Separating features and labels
X = diabetes.drop(columns='Outcome', axis=1)
Y = diabetes['Outcome']

# Initialize lists to store bootstrapped model results
accuracy_scores = []
f1_scores = []

n_bootstrap = 10  # Number of bootstrapped samples

for i in range(n_bootstrap):
    # Generate a bootstrapped sample
    bootstrap_indices = np.random.choice(range(len(X)), size=len(X), replace=True)
    X_bootstrap = X.iloc[bootstrap_indices]
    Y_bootstrap = Y.iloc[bootstrap_indices]

    # Train-test split on the bootstrapped sample
    X_train, X_test, Y_train, Y_test = train_test_split(X_bootstrap, Y_bootstrap,
                                                        test_size=0.2, stratify=Y_bootstrap, random_state=3)

    # Standardize the data
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train_std = scaler.transform(X_train)
    X_test_std = scaler.transform(X_test)

    # Create and train the KNN model
    model = KNeighborsClassifier(n_neighbors=11)
    model.fit(X_train_std, Y_train)

    # Evaluate the model on the test set
    Y_test_predict = model.predict(X_test_std)
    accuracy = accuracy_score(Y_test, Y_test_predict)
    f1 = f1_score(Y_test, Y_test_predict, average='macro')

    # Store the evaluation results
    accuracy_scores.append(accuracy)
    f1_scores.append(f1)

# Calculate the average evaluation scores over the bootstrapped models
avg_accuracy = np.mean(accuracy_scores)
avg_f1 = np.mean(f1_scores)

print("Average Accuracy:", avg_accuracy)
print("Average F1 Score:", avg_f1)
```
