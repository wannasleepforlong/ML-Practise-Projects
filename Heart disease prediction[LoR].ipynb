```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
heart_df=pd.read_csv('/content/heart_disease_data.csv')
heart_df.head()
```

```python
heart_df.describe
```

```python
#checking for missing values
heart_df.isnull().sum()
```

```python
heart_df.info()
```

```python
X=heart_df.drop(columns='target',axis=1)
Y=heart_df['target']
```

```python
#Splitting
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
#Scaling
#cant scale cos one of it is string:((
```

```python
model=LogisticRegression(solver='lbfgs', max_iter=1000)#else error because can't converge
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(X_train_prediction, Y_train)
print(train_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_accuract=accuracy_score(X_test_prediction,Y_test)
```
