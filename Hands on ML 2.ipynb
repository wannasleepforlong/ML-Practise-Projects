```python
from tensorflow import keras
(X_train,Y_train),(X_test,Y_test)=keras.datasets.mnist.load_data()
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
classifier=SGDClassifier()
Y_train_5=(Y_train==5)
Y_test_5=(Y_test==5)
#makes =5 true, rest as false
```

```python
print(X_train[0])
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
classifier.fit(X_train.reshape(-1, 28 * 28), Y_train_5)
```

```python
Y_pred_5=classifier.predict(X_test.reshape(-1,28*28))
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(Y_pred_5,Y_test_5)
```
