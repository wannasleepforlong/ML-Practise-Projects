```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import nltk
```

```python
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
mail_data=pd.read_csv('/content/mail_data.csv')
mail_data.head()
```

```python
mail_data.isnull().sum()
```

```python
mail_df=mail_data.where(pd.notnull(mail_data),'')
#can handle missing values if any and is useful in text based analySis
```

```python
#label encoding
#we do not use label encoder because our df isnt large
mail_df.loc[mail_df['Category']=='spam','Category']=0
#locate where mail_df['Category']=='spam' is true and replace the Category value at that row to 0
mail_df.loc[mail_df['Category']=='ham','Category']=1
```

```python
mail_df['Category'].value_counts()
```

```python
X=mail_df['Message']
Y=mail_df['Category']
```

```python
#Splitting
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
#Feature extraction
vectorizer=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
#The default min_df is 1, which means "ignore terms that appear in less than 1 document"
#it will also convert all words to lowercase
```

```python
X_train_features=vectorizer.fit_transform(X_train)
X_test_features=vectorizer.transform(X_test)
```

```python
print(X_train_features)
```

```python
Y_train=Y_train.astype('int')
#converting obj to str
Y_test=Y_test.astype('int')
```

```python
#Training
model=LogisticRegression()
model.fit(X_train_features,Y_train)
#predicting accuracy
X_train_pred=model.predict(X_train_features)
train_accu=accuracy_score(X_train_pred,Y_train)
print(train_accu)
```

```python
X_test_pred=model.predict(X_test_features)
test_accu=accuracy_score(X_test_pred,Y_test)
print(test_accu)
```
